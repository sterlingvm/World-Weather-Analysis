```python
# Import the random module
import random
```

```python
# Generate a random integer value between two parameters
random.randint(-90,90)
```

```python
# Generate a random float value between 0 & 1
random.random()
```

```python
# Generate a random integer
random_number = random.randint(-90,89) + random.random()
random_number
```

```python
# Use a while loop to iterate and generate 10 random int + <-1-float values
x = 1
latitudes = []
while x < 11:
    rand_lat = random.randint(-90,89) + random.random()
    latitudes.append(rand_lat)
    x += 1
    
latitudes
```

```python
# Generate a random integer based on a sequence with x-distanced steps
random.randrange(-90,90, step=2)
```

```python
# Generate a random float point value within a two-parameter range
random.uniform(-90,90)
```

```python
# Import timeit | This is a module that tells you how long a piece of code takes to run !
import timeit
```

```python
%timeit np.random.uniform(-90, 90, size=1500)
```

```python
def latitudes(size):
    latitudes = []
    x = 0
    while x < size:
        random_lat = random.randint(-90,90) + random.random()
        latitudes.append(random_lat)
        x += 1
    return latitudes

# Call the function with 1500.
%timeit latitudes(1500)
```
